```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data.csv')
df = df.drop('id', 1)
# преобразование целевой переменной в булевы значения
df = df.replace('M', 1, regex=True) # M = ill = true
df = df.replace('B', 0, regex=True) # B = not ill = false
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
Corr=df.corr()
```

```python
plt.figure(figsize=(30, 30))
ax = sns.heatmap(Corr, annot=True, fmt='.2f')
ax.set_ylim(32, 0) # fix cutted edges
ax
# на тепловой карте видно, с какими фичами коррелирует целевая переменная - diagnosis
```

```python
sns.boxplot(x="diagnosis", y="radius_mean", data=df) # пример высокой корреляции
```

```python
sns.regplot(x=df["concave points_worst"], y=df["diagnosis"], fit_reg=False) # пример высокой корреляции
```

```python
sns.boxplot(x="diagnosis", y="perimeter_worst", data=df) # пример высокой корреляции
```

```python
sns.regplot(x=df["texture_se"], y=df["diagnosis"], fit_reg=False) # пример низкой корреляции
```

```python
sns.boxplot(x="diagnosis", y="symmetry_se", data=df) # пример низкой корреляции
```

```python
# В целом понятно, какие фичи больше всего влияют на целевую переменную
```
